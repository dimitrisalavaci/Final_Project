```python
home_win = (df['home_team_win'] == '1').sum()
df['home_team_total_wins'] = [home_win if df.loc[i, 'home_team_win'] == '1' else 0 for i in range(len(df))]
```

```python
away_win = (df['home_team_win'] == '0').sum()
df['away_team_total_wins'] = [away_win if df.loc[i, 'home_team_win'] == '0' else 0 for i in range(len(df))]
df
```

```python
# Load the train and test datasets

train_df = pd.read_csv(r"C:\Users\Dimitri Salavaci\Downloads\nbatrain.csv")
test_df = pd.read_csv(r"C:\Users\Dimitri Salavaci\Downloads\nbatest.csv")
```

```python
# Separate the target variable from the input features
X_train = train_df.drop('home_team_win', axis=1)
y_train = train_df['home_team_win']
X_test = test_df.drop('home_team_win', axis=1)
y_test = test_df['home_team_win']
```

```python
# import libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
```

```python
pca = PCA(n_components=9)
pca.fit(X_train_scaled)
X_train_pca = pca.transform(X_train_scaled)
```

```python
model = LogisticRegression()
model.fit(X_train_pca, y_train)
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_pca = pca.transform(X_test_scaled)
```

```python
y_pred = model.predict(X_test_pca)
acc = accuracy_score(y_test, y_pred)
print(f'Test set accuracy: {acc}')
```

```python
import pickle
# save the model to disk
filename = 'final_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'penalty' : ['l1','l2'], 
    'C'       : np.logspace(-3,3,7),
    'solver'  : ['newton-cg', 'lbfgs', 'liblinear'],
}
```

```python
logreg = LogisticRegression()
new_model = GridSearchCV(logreg,                    
                   param_grid = parameters,   
                   scoring='accuracy',        
                   cv=10)  
```

```python
new_model.fit(X_train,y_train)
```

```python
print("Tuned Hyperparameters :", new_model.best_params_)
print("Accuracy :",new_model.best_score_)
```

```python
logreg = LogisticRegression(C = 0.001, 
                            penalty = 'l1', 
                            solver = 'liblinear')
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print("Accuracy:",logreg.score(X_test, y_test))
```

```python
# save the model to disk
filename2 = 'final_model.sav'
pickle.dump(logreg, open(filename2, 'wb'))
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline(steps=[('scaling', StandardScaler()),
                           ('pca', PCA(n_components=5)),
                           ('classifier', LogisticRegression())])
pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'Test set accuracy: {acc}')
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```
